```python
# Importing all the necessary files
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# --- 0. Loading Data ---
# We upload the college.xlsx into a dataframe named 'df'.
df = pd.read_excel('https://raw.githubusercontent.com/mayuresh-tungare/c4m2-graded/e363c13d8b8ac802b4720a48ee3473ba46f03191/college.xlsx')
print(df.info)
print(df.columns)
```

```python
# --- 1. Remove the university/ college name column ---
# As we already loaded the data without index_col=0, we will drop the first (blank) column as Unnamed: 0 and then check this by running print(df.columns):
df = df.drop(columns=['Unnamed: 0'])
print(df.info)
print(df.columns)
```

```python
# --- 2. Descriptive statistics on numeric columns ---
# Using .describe() and rounding to 2 decimal places using .round()
desc_stats = df.describe().round(2)
print("--- Descriptive Statistics ---")
print(desc_stats)
print("\n")
```

```python
# --- 3. One-hot encode the 'Private' variable ---
# Using pandas.get_dummies() and dropping the first encoded column to avoid multicollinearity
df = pd.get_dummies(df, columns=['Private'], drop_first=True)
print("--- One-Hot Encoded Data ---")
print(df.head())
print("\n")
```

```python
# --- 4. Define Target Variable & Features ---
# 'Grad.Rate' is the target variable (y). The rest are features (X).
# Note: As requested, no normalization or scaling is applied to X.
X = df.drop(columns=['Grad.Rate'])
y = df['Grad.Rate']
print(X)
print(y)
```

```python
# --- 5. Train/Test Split ---
# 70:30 split, setting random_state to 42 for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
# --- 6. Fit the Model ---
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# --- 7. Output Model Coefficients ---
# Converting coefficients to a Pandas Series to use the .round(2) method easily
coefficients = pd.Series(model.coef_, index=X.columns).round(2)
print("--- Model Coefficients ---")
print(coefficients)
print("\n")
```

```python
# --- 8. Obtain Predictions ---
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# --- 9. Generate and Round Metrics ---
# Using round() function to round sklearn.metrics outputs to 2 decimal places
metrics = {
    "Training Set": {
        "R²": round(r2_score(y_train, y_train_pred), 2),
        "MSE": round(mean_squared_error(y_train, y_train_pred), 2),
        "MAE": round(mean_absolute_error(y_train, y_train_pred), 2)
    },
    "Testing Set": {
        "R²": round(r2_score(y_test, y_test_pred), 2),
        "MSE": round(mean_squared_error(y_test, y_test_pred), 2),
        "MAE": round(mean_absolute_error(y_test, y_test_pred), 2)
    }
}

print("--- Performance Metrics ---")
for dataset, scores in metrics.items():
    print(f"{dataset}:")
    print(f"  R²  : {scores['R²']}")
    print(f"  MSE : {scores['MSE']}")
    print(f"  MAE : {scores['MAE']}")
```
